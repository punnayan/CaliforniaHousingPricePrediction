```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

```python
housing.keys()
```

```python
print(housing.DESCR)
```

```python
dataset = pd.DataFrame(housing.data,columns = housing.feature_names)
```

```python
dataset.head()
```

```python
dataset['Price'] = housing.target
```

```python
dataset.head()
```

```python
## check missing value
```

```python
dataset.isnull().sum()
```

```python
## exploring data analysis
## correlation
dataset.corr()
```

```python
import seaborn as sns 
sns.pairplot(dataset)
```

```python
plt.scatter(dataset['MedInc'],dataset['Price'])
plt.xlabel("Median income in block group")
plt.ylabel("Price")
```

```python
import seaborn as sns
sns.regplot(x="Population" , y = "Price",data = dataset)
```

```python
sns.regplot(x="Longitude" , y = "Price",data = dataset)
```

```python
sns.regplot(x="HouseAge" , y = "Price",data = dataset)
```

```python
## Independent and Dependent features 
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size =0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
## standardise the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
##model training
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
##print the coefficient and the intercept
print(regression.coef_)
```

```python
##on which parameter the has been trained
regression.get_params()
```

```python
## prediction with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for the prediction
plt.scatter(Y_test,reg_pred)
```

```python
## residuals = errors
residuals = Y_test-reg_pred
```

```python
residuals
```

```python
##plot this residuals
sns.displot(residuals,kind="kde")
```

```python
#scatter plot with respect to prediction and residuals
## uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(Y_test,reg_pred))
print(mean_squared_error(Y_test,reg_pred))
```

```python
## R square and adjusted R square
from sklearn.metrics import r2_score
score = r2_score(Y_test,reg_pred)
print(score)
```

```python
#display adjusted r-square
1-(1-score)*(len(Y_test)-1)/(len(Y_test)-X_test.shape[1]-1)
```
